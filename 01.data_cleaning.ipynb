```python
import pandas as pd

# Load your dataset
#df = pd.read_csv("your_dataset.csv")  # Replace with your actual file path

# Select important columns
selected_columns = [
    'UTC.time',
    'Local.date',
    'Region',
    'Hour',
    'D',  # Actual Electricity Demand
    'NG',  # Net Generation
    'TI',  # Total Interchange
    'DF',  # Forecasted Demand
    'NG..COL', 'NG..NG', 'NG..OIL', 'NG..NUC', 'NG..WAT', 'NG..SUN', 'NG..WND',
    'CO2.Factor..COL', 'CO2.Factor..NG', 'CO2.Factor..OIL',
    'CO2.Emissions..COL', 'CO2.Emissions..NG', 'CO2.Emissions..OIL',
    'CO2.Emissions.Generated',
    'CO2.Emissions.Intensity.for.Generated.Electricity',
    'CO2.Emissions.Intensity.for.Consumed.Electricity'
]

# Filter the dataset
data = df[selected_columns]

# Show the first few rows of the filtered dataset
data
```

```python
data.rename(columns={
    'UTC.time': 'UTC_Time',
    'Local.date': 'Date',
    'Region': 'Region',
    'Hour': 'Hour',
    'D': 'Demand',
    'NG': 'Net_Generation',
    'TI': 'Total_Interchange',
    'DF': 'Forecasted_Demand',
    
    # Energy generation by source
    'NG..COL': 'Coal_Gen',
    'NG..NG': 'Gas_Gen',
    'NG..OIL': 'Oil_Gen',
    'NG..NUC': 'Nuclear_Gen',
    'NG..WAT': 'Hydro_Gen',
    'NG..SUN': 'Solar_Gen',
    'NG..WND': 'Wind_Gen',
    
    # CO2 Emission Factors
    'CO2.Factor..COL': 'CO2_Factor_Coal',
    'CO2.Factor..NG': 'CO2_Factor_Gas',
    'CO2.Factor..OIL': 'CO2_Factor_Oil',
    
    # CO2 Emissions
    'CO2.Emissions..COL': 'CO2_Emissions_Coal',
    'CO2.Emissions..NG': 'CO2_Emissions_Gas',
    'CO2.Emissions..OIL': 'CO2_Emissions_Oil',
    'CO2.Emissions.Generated': 'CO2_Total_Emissions',
    
    # CO2 Intensity
    'CO2.Emissions.Intensity.for.Generated.Electricity': 'CO2_Intensity_Generated',
    'CO2.Emissions.Intensity.for.Consumed.Electricity': 'CO2_Intensity_Consumed'
}, inplace=True)
```

```python
# Convert mixed-format dates correctly
data['Date'] = pd.to_datetime(data['Date'], format='mixed', dayfirst=True, errors='coerce')

# Check for rows that failed to convert
invalid_dates = data[data['Date'].isna()]
print(f"Number of invalid dates: {len(invalid_dates)}")

# Optional: Drop rows with invalid dates
data = data.dropna(subset=['Date'])

# View sampledata
data
```

```python
data.isna().sum()
```

```python
data.drop(columns=['Oil_Gen', 'Wind_Gen', 'CO2_Factor_Oil', 'CO2_Emissions_Oil'], inplace=True)
data
```

```python
# Fill generation-related gaps with 0 (likely non-generated)
gen_cols = ['Gas_Gen', 'Hydro_Gen', 'Solar_Gen', 'Nuclear_Gen']
data[gen_cols] = data[gen_cols].fillna(0)

# CO2 factors: fill with column means
data['CO2_Factor_Coal'] = data['CO2_Factor_Coal'].fillna(data['CO2_Factor_Coal'].mean())
data['CO2_Factor_Gas'] = data['CO2_Factor_Gas'].fillna(data['CO2_Factor_Gas'].mean())

# CO2 emissions fill with 0 (non-generation = no emission)
data[['CO2_Emissions_Coal', 'CO2_Emissions_Gas']] = data[['CO2_Emissions_Coal', 'CO2_Emissions_Gas']].fillna(0)
data
```

```python
# Sort data for interpolation
data = data.sort_values(by=['Region', 'Date', 'Hour'])

# Interpolate based on time (good for time series)
interp_cols = [
    'Demand', 'Net_Generation', 'Total_Interchange', 'Forecasted_Demand',
    'CO2_Total_Emissions', 'CO2_Intensity_Generated', 'CO2_Intensity_Consumed'
]
data[interp_cols] = data[interp_cols].interpolate(method='linear', limit_direction='both')
data
```

```python
data = data.dropna()
data
```

```python
numeric_cols = ['Demand','Net_Generation','Total_Interchange','Forecasted_Demand','Coal_Gen',
                'Gas_Gen','Nuclear_Gen','Hydro_Gen','Solar_Gen','CO2_Factor_Coal','CO2_Factor_Gas',
                'CO2_Emissions_Coal','CO2_Emissions_Gas','CO2_Total_Emissions','CO2_Intensity_Generated','CO2_Intensity_Consumed']


# Round all numeric columns to 2 decimal places
data[numeric_cols] = data[numeric_cols].round(2)
data
```
